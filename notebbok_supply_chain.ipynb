# Projet Supply Chain – Planification de projet

## Sujet 05 – Montage d'un film

## 1. Objectif

L’objectif de ce projet est de déterminer l’organisation des différentes tâches nécessaires à la réalisation d’un film.

Nous cherchons à :
- calculer les dates de début et de fin de chaque tâche
- déterminer la durée totale du projet
- identifier les tâches critiques

## 2. Compréhension du problème

Le projet est composé de plusieurs tâches (de A à L).

Chaque tâche possède :
- une durée
- des contraintes (appelées prérequis)

Une tâche ne peut commencer que lorsque toutes ses contraintes sont respectées.

Par exemple :
- certaines tâches doivent attendre la fin d’une autre tâche
- d’autres doivent attendre plusieurs tâches
- certaines nécessitent un délai supplémentaire après une tâche

## 3. Méthode

Pour résoudre ce problème, nous calculons les dates de début et de fin au plus tôt pour chaque tâche.

Les règles utilisées sont :

- Début = moment où toutes les contraintes sont respectées
- Fin = début + durée

Lorsque plusieurs tâches sont nécessaires, on attend la plus longue.

## 4. Calcul des dates au plus tôt

### Tâche A

Aucun prérequis.

Début = 0  
Fin = 0 + 30 = 30

### Tâche B

B commence 15 jours après la fin de A.

A finit à 30  
Début = 30 + 15 = 45  
Fin = 45 + 12 = 57


### Tâche C

C commence 20 jours après la fin de A.

Début = 30 + 20 = 50  
Fin = 50 + 8 = 58

### Tâche D

D dépend de A et C.

A = 30  
C = 58  

Début = max(30, 58) = 58  
Fin = 58 + 4 = 62

### Tâche E

E dépend de C et D.

C = 58  
D = 62  

Début = max(58, 62) = 62  
Fin = 62 + 7 = 69

### Tâche F

F dépend de A, B, C et D.

A = 30  
B = 57  
C = 58  
D = 62  

Début = max(30, 57, 58, 62) = 62  
Fin = 62 + 10 = 72

### Tâche G

G dépend de D, E et F.

D = 62  
E = 69  
F = 72  

Début = max(62, 69, 72) = 72  
Fin = 72 + 12 = 84

### Tâche H

H dépend de F et G.

F = 72  
G = 84  

Début = max(72, 84) = 84  
Fin = 84 + 3 = 87


### Tâche I

I dépend de H.

Début = 87  
Fin = 87 + 14 = 101

### Tâche J

J dépend de :
- début de I + 3 jours
- fin de H

Début I = 87 → 87 + 3 = 90  
H = 87  

Début = max(90, 87) = 90  
Fin = 90 + 7 = 97

### Tâche K

K dépend de I et J.

I = 101  
J = 97  

Début = max(101, 97) = 101  
Fin = 101 + 6 = 107

### Tâche L

L dépend de la fin de K + 2 jours.

K = 107 → 107 + 2 = 109  

Début = 109  
Fin = 109 + 1 = 110

## 5. Durée totale du projet

La durée totale du projet est de :

**110 jours**

## 6. Chemin critique

Le chemin critique correspond aux tâches qui déterminent la durée du projet.

Chemin critique :

A → C → D → F → G → H → I → K → L

## 7. Interprétation

Les tâches critiques sont celles qui n’ont aucune marge.

Cela signifie que si l’une de ces tâches est en retard, alors l’ensemble du projet sera retardé.

À l’inverse, certaines tâches comme B ou E ne sont pas critiques, car elles se terminent avant d’autres tâches plus longues.

Par exemple, la tâche E se termine avant F, donc elle ne retarde pas le projet.

Ainsi, seules les tâches les plus longues à chaque étape influencent la durée totale.

## 8. Généralisation du problème

Le problème étudié précédemment (montage d’un film) est un cas particulier d’un problème plus général d’ordonnancement de tâches.

De manière générale, un projet peut être représenté comme un ensemble de tâches.

Chaque tâche possède :
- une durée
- des dépendances vis-à-vis d’autres tâches

Une dépendance signifie qu’une tâche ne peut commencer qu’après le début ou la fin d’une autre tâche, éventuellement avec un décalage temporel.

Ainsi, une tâche peut être décrite par :
- un identifiant
- une durée
- une liste de dépendances

Ce modèle est général et peut s’appliquer à différents domaines comme la gestion de projet, la supply chain, la production ou le développement informatique.

```python
# Exemple de modélisation générale

general_tasks = {
    "T1": {
        "duration": 5,
        "dependencies": []
    },
    "T2": {
        "duration": 3,
        "dependencies": [
            {"task": "T1", "type": "finish", "lag": 2}
        ]
    },
    "T3": {
        "duration": 4,
        "dependencies": [
            {"task": "T1", "type": "finish", "lag": 0},
            {"task": "T2", "type": "finish", "lag": 0}
        ]
    },
    "T4": {
        "duration": 2,
        "dependencies": [
            {"task": "T2", "type": "start", "lag": 1}
        ]
    }
}

general_tasks
```
